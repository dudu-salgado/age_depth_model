```python
import numpy as np
import matplotlib.pyplot as plt

# Known depths and ages
known_depths = [0, 1.29, 2.11, 2.73, 3.4]  # Depths in meters
known_age_central = [0, 5754, 9927, 13287, 18277]  # Central ages in years (C14)
known_age_uncertainties = [0, 164, 320, 315, 408]  # Age uncertainties in years (C14)

# Function to interpolate ages
def interpolate_ages(depth, known_depths, known_ages):
    return np.interp(depth, known_depths, known_ages)

# Create a depth grid for the model
model_depths = np.linspace(0, 3.5, 100)

# Calculate corresponding central ages
model_central_ages = interpolate_ages(model_depths, known_depths, known_age_central)

# Calculate corresponding uncertainties
model_age_uncertainties = interpolate_ages(model_depths, known_depths, known_age_uncertainties)

# Invert the y-axis
model_depths_inverted = 3.5 - model_depths

# Plot the age-depth model with uncertainty bands
plt.figure(figsize=(8, 6))
plt.plot(model_central_ages, model_depths_inverted, color='black', label='Age-Depth Model')
plt.fill_betweenx(model_depths_inverted, model_central_ages - model_age_uncertainties, model_central_ages + model_age_uncertainties, color='gray', alpha=0.3, label='Uncertainty')
plt.scatter(known_age_central, 3.5 - np.array(known_depths), color='red', label='Known C14 Ages')
plt.xlabel('Age (C14 years)')
plt.ylabel('Depth (meters)')
plt.title('Age-Depth Model with Uncertainty Bands')
plt.legend()
plt.gca().invert_yaxis()  # Invert the y-axis
plt.grid(True)
plt.show()

# Save the graph as a PNG file
plt.savefig('age_depth_model1.png', bbox_inches='tight', dpi=300)


# Save the graph as an SVG file for Illustrator
plt.savefig('age_depth_model.svg', format='svg', bbox_inches='tight', transparent=True)

# Show the graph
plt.show()
```
