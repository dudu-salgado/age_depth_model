```python
import numpy as np
import matplotlib.pyplot as plt

# Profundidades e idades conhecidas
profundidades_conhecidas = [0, 1.29, 2.11, 2.73, 3.4]  # Profundidades em metros
idades_centrais_conhecidas = [0, 5754, 9927, 13287, 18277]  # Idades centrais em anos (C14)
incertezas_conhecidas = [0, 164, 320, 315, 408]  # Faixas de incerteza em anos (C14)

# Função para interpolar idades
def interpolar_idades(profundidade, profundidades_conhecidas, idades_conhecidas):
    return np.interp(profundidade, profundidades_conhecidas, idades_conhecidas)

# Criar uma grade de profundidades para o modelo
profundidades_modelo = np.linspace(0, 3.5, 100)

# Calcular idades centrais correspondentes
idades_modelo_centrais = interpolar_idades(profundidades_modelo, profundidades_conhecidas, idades_centrais_conhecidas)

# Calcular incertezas correspondentes
idades_modelo_incertezas = interpolar_idades(profundidades_modelo, profundidades_conhecidas, incertezas_conhecidas)

# Inverter o eixo y
profundidades_modelo_invertido = 3.5 - profundidades_modelo

# Plotar o age-depth model com faixas de incerteza
plt.figure(figsize=(8, 6))
plt.plot(idades_modelo_centrais, profundidades_modelo_invertido, color='black', label='Age-Depth Model')
plt.fill_betweenx(profundidades_modelo_invertido, idades_modelo_centrais - idades_modelo_incertezas, idades_modelo_centrais + idades_modelo_incertezas, color='gray', alpha=0.3, label='Incerteza')
plt.scatter(idades_centrais_conhecidas, 3.5 - np.array(profundidades_conhecidas), color='red', label='Idades C14 conhecidas')
plt.xlabel('Idade (anos C14)')
plt.ylabel('Profundidade (metros)')
plt.title('Age-Depth Model com Faixas de Incerteza')
plt.legend()
plt.gca().invert_yaxis()  # Inverter o eixo y
plt.grid(True)
plt.show()

# Plotar o age-depth model com faixas de incerteza
plt.figure(figsize=(8, 6))
plt.plot(idades_modelo_centrais, profundidades_modelo_invertido, color='black', label='Age-Depth Model')
plt.fill_betweenx(profundidades_modelo_invertido, idades_modelo_centrais - idades_modelo_incertezas, idades_modelo_centrais + idades_modelo_incertezas, color='gray', alpha=0.3, label='Incerteza')
plt.scatter(idades_centrais_conhecidas, 3.5 - np.array(profundidades_conhecidas), color='red', label='Idades C14 conhecidas')
plt.xlabel('Idade (anos C14)')
plt.ylabel('Profundidade (metros)')
plt.title('Age-Depth Model com Faixas de Incerteza')
plt.legend()
plt.gca().invert_yaxis()  # Inverter o eixo y
plt.grid(True)

# Salvar o gráfico como um arquivo PNG
plt.savefig('age_depth_model1.png', bbox_inches='tight', dpi=300)

# Salvar o gráfico como um arquivo PNG para excel

plt.savefig('age_depth_model3.emf', format='emf', bbox_inches='tight')

# Salvar o gráfico como um arquivo para Illustrator
plt.savefig('age_depth_model.svg', format='svg', bbox_inches='tight', transparent=True)


# Mostrar o gráfico
plt.show()
```
